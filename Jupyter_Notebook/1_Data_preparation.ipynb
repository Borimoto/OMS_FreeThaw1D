## How to write the time series for OMS

OMS time series has to be formatted as:

```python
from IPython.display import Image
Image(filename='image/OMS_time_series.png',width=800, height=400)
```

- The first line contains some control string
- The second line the date of creation
- The third line general info
- The fourth line header used by the component
- The fifth line start with a string that is one of the input of the OMS component, `org.jgrasstools.gears.io.timedependent.OmsTimeSeriesIteratorReader.idfield`. The value that follows are the header of the columns containing the data. These values may be the ID of meteo stations, or generic values and in this case columns must be numbered from zero 
- The sixth line contains some st
- From the seventh line and below we have the fisrt column empty,  the second contains the date and from the third onwards there are values.

This file can be obtained with the function `gf_io.write_OMS_timeseries`.


```python
df_noformatted = pd.read_csv(work_directory_path+'/data/Timeseries/data.csv', low_memory = False)
df_noformatted.head()
```

```python
gf_io.write_OMS_timeseries(df_noformatted.iloc[:,1:],'02-01-2020 00:00','1H',work_directory_path+'/data/Timeseries/data_formatted.csv')
```
