## Abstract

A common problem of thermal models for permafrost, and for freezing soil, regards the treatment of the nonlinear behaviour of the energy and enthalpy as function of temperature, water content and other governing variables.
As a result, models may fail to converge or to obtain the correct solution. 
In this paper we use a recent algorithm, called nested Newton-Casulli-Zanolli (NCZ), previously applied in other problems, to solve neatly these integration issues.
We review  first the most common approaches used to model freezing soil, focusing in particular on the so called fixed-grid method.
We apply the algorithm, then, to a finite volume discretization of the energy-enthalpy formulation and the novelty regards the application of NCZ and remark the importance to use a so-called conservative formulation of the governing equation, an aspect which is usually overlooked by researchers in the field.
Using our formulation and the NCZ algorithm, the convergence of the solver is guaranteed for any time step size. This is a key feature since the integration time step can then be chosen according to the time scale of the processes to study, from seconds to days.
The model is successfully tested for large integration time step against the Neumann analytical solution, both for the freezing of water and for the meltging of ice, and the Lunardini analytical solution.
With a long-term synthetic one dimensional experiment covering 100 years we show that the numerical solution obtained with an integration time step of ten days is in close agreement with that obtained with an integration time step of one hour, and the convergence is still assured.

# Descriptions for general Audience

Freezing and thawing of soils affect a wide range of biogeochemical and hydrological ([Walvoord and Kurylyk, 2016)](https://doi.org/10.2136/vzj2016.01.0010); [Schuur
et al., 2015](https://doi.org/10.1038/nature14338)) processes and interact with engineered structures in cold regions. This foundational importance together with the vast extent of cold regions globally makes the simulation of freezing and thawing soil an important and well-researched topic ([Streletskiy et al., 2019](https://doi.org/10.1088/1748-9326/aaf5e6); [Walvoord and Kurylyk, 2016](https://doi.org/10.2136/vzj2016.01.0010); [Harris et al., 2009](https://doi.org/10.1016/j.earscirev.2008.12.002)). The interest in understanding and anticipating the consequences of permafrost thaw driven by anthropogenic climate change has added additional urgency.

Studying frozen soil can be done at different timescales. At short timescale, the latent heat of fusion of water determines a time-lag in the cooling and warming of the soil thus affecting the surface heat fluxes. Studies have shown that proper frozen soil scheme help improve land surface and climate model simulation ([Viterbo et al., 1999](https://doi.org/10.1002/qj.49712555904); [Smirnova et al., 2000](https://doi.org/10.1029/1999JD901047)). 
On the other hand, permafrost evolution is characterized by a long timescale and its evolution is manly controlled by climate change. Permafrost degradation has significant impact on the ecosystems and on the social and economic life in cold regions ([Bao et al., 2016](https://doi.org/10.1002/2015JD024451))
Therefore, it is desirable to have a numerical model that can efficiently simulate frozen soil at both short timescale and long timescale.
At present, a common problem of physical-based frozen soil model concerns the treatment of the nonlinear behavior of the energy and enthalpy as function of temperature, water content and other governing variables. Because of this nonlinearity, the model may fail to compute the correct solution, or a short time step must be chosen leading to an increase of the computational cost.

`FreThaw1D` model solves the so called enthalpy - or conservative - formulation. The novelty regards the use of the nested Newton-Casulli-Zanolli (NCZ) algorithm, ([Casulli and Zanolli, 2010](https://doi.org/10.1137/100786320)) to linearize the nonlinear system resulting from the approximation of the governing equation. Using the enthalpy formulation and the NCZ algorithm the convergence of the solver is guarateed for any time step size. This is a key feature since the integration time step can be choosen accordingly to the time scale of the processes to study, from seconds to days.

## Inputs

- Computational grid data
- Boundary condition data
- Simulation's control parameters

The computational grid data in input include:
- The name of the [NetCDF file](https://en.wikipedia.org/wiki/NetCDF) containing the data of the computational grid.

The boundary condition data include the names of the files (.csv format) containing:
-  the surface boundary condition, if Dirichlet \[K\] if Neumann \[ W m$^{-2}$\]
-  the bottom boundary condition, if Dirichlet \[K\] if Neumann \[ W m$^{-2}$\]

A time series of 1 or 0 to select which dates to save as output: 1 to save, 0 not.

The Control parameters include: 

for each of the boundary condition data file above:

- The id of the location under analysis
- The start date of simulation
- The end date of simulation
- The time step of simulation
- The novalue, it is necessary $-9999$.

and:

- tTimestep
- timeDelta
- nestedNewton
- newtonTolerance
- delta
- picardIteration
- writeFrequency
- topBCType
- bottomBCType
- sfccModel
- stateEquationModel
- soilThermalConductivityModel
- interfaceThermalConductivityModel

### Detailed inputs Description

####  tTimeStep

Type `double`

Time step of the boundary condition timeseries \[ s \]. 

####  timeDelta

Type `double`

Integration time step \[ s \]. 


#### nestedNewton

Type `int`

Type of the algorithm used to linearized the system of equation. The default value is 1 which corresponds to the nested Newton-Casulli-Zanolli algorithm, 0 corresponds the Newton-Raphson algorithm.

#### newtonTolerance

Type `double`

Tolerance of the algorithm used to linearized the system of equation.  


#### delta

Type `double`

Paramenter used to improve the convergence of the Newton-Raphson algorithm. Defeault value is 1 and correspond to the Newton-Raphson algorithm, if <1 the the so called globally convergent Newton algorithm is used. The range of this value is \]0,1\]. This is used only if nestedNewton is set to 0.  

#### picardIteration

Type `int`

Number of Picard iteration that can be used to update the thermal conductivity. This is a semi-impicit method therefore the thermal conductivity is computed with the solution - temperature - of the previous time step. It is worth to underline that this variable does not controll in any manner the convergence of the numerical solution. It could improve the modelization of the thermal conductivity. Range $\leq$1. 

#### writeFrequency

Type `int`

Number of time steps every which the output are saved on the disk. Range $\leq$1. 

#### topBCType

Type `String`

Type of the top boundary condition \[Top Dirichlet, Top Neumann\]

#### bottomBCType

Type `String`

Type of the bottom boundary condition \[Bottom Dirichlet, Bottom Neumann\]

#### sfccModel

Type `String`

The name of the SFCC model used in the simulation. It is added as attribute in the NetCDF file. \[DallAmico, Lunardini, McKenzie, None\]

#### stateEquationModel

Type `String`
 
The name of the state equation model used in the simulation. For instance one can model the phase change of pure water, or the thermal regime of the ground: the enthalpy fuction - state equation - is different. \[Water, Lunardini, Soil\]

### soilThermalConductivityModel

Type `String`

the name of the model used to define the thermal conductivity of the medium. It is added as attribute in the NetCDF file. \[Johansen, water, Luanrdini\]

#### interfaceThermalConductivityModel

Type `String`

The name of the model used to evaluate the thermal conductivity at the interface of the adjacent control volumes. \[Arithmetic mean, max, min, Harmonic mean, Geometric mean\]

#### myVariables

Type: `ArrayList<double[]>`

This variable contains all the output of the current time step.  The output are:
- temperature [K], 
- water content [-],
- ice content [-],
- enthalpy (i.e. internal energy) [J],
- error 
- heat flux through the surface boundary [W m-2]
- heat flux through the bottom boundary [W m-2]

### Boundary condition
Solving a partial differential equation (PDE) require the definition of the boundary condition at the border of the domain. This is a 1D model thus we need to prescribe two boundary condtions: one at the top of the domain and one at the bottom of the domain. At the moment two types of boundary condition are available:
- [Dirichlet](https://en.wikipedia.org/wiki/Dirichlet_boundary_condition)
- [Neumann](https://en.wikipedia.org/wiki/Neumann_boundary_condition)

To implement these boundary condition we adopt the `Simple factory design pattern`

### State equation
The function describing the enthalpy function depends on the problem to study. This numerical model has been primarly developed to simulate the ground thermal regime but it can be used to simulate freezing and thawing of pure water as well. Thus the definition of the enthalpy function is not unique. Again the `Simple Factory pattern` has been adopted.

Considering the problem of frozen soil it is worth to note that the enthalpy function is defined as a weighted average of the constituents of the soil: soil particels, liquid water, and ice. In general the enthalpy contribute of air is neglected. The liquid water content is a function of temperature and is described by the so called Soil Freezing Characteristic Curves (SFCC). Accordingly to the OOP [open-closed paradigm](https://en.wikipedia.org/wiki/Open%E2%80%93closed_principle#:~:text=In%20object%2Doriented%20programming%2C%20the,without%20modifying%20its%20source%20code.), the implementation of the state equation made use of another design patter: the `decorator pattern`. 



````python
Image(filename='image/UML_stateequation.jpg',width=1500, height=1500/1.618)
````

The abstract class `StateEquation` has been developed in another Java project, as well as the abstract class `Rheology`. This Java project is imported as a .jar file.
This choice is motivated by the fact the definition of the state equation is a cross-cutting problem. Moreover the state equation in general may depend on the definition of other functions, for instance the SFCC, that we called as rheology function. Let us think to another well known equation in Hydrology such as the Richards' equation. Here when we solve the PDE the state equation is represented by the integral over the control volume of the Soil Water Retention Curve: the state equation is the water volume function, that plays the same role of the enthalpy function, and the SWRC corresponds to the SFCC.


Besides the abstract classes `StateEquation` and `Rheology` another abstract class has been implemented: the `ClosureEquation` class. This class is meant to model the closure equation of the problem. It is again an abstract class, we use again the `Simple Factory pattern`, since the closure equation depends on the problem under scrutiny, and for the same problem it is possible that in literature different models have been presented.
